```python
import matplotlib.pyplot as plt
import numpy as np

# Data
experiments = ['Experiment 1', 'Experiment 2', 'Experiment 3', 'Experiment 4', 'Experiment 5']
num_reviews = [1000, 2000, 3000, 4000, 5000]
accuracies = [0.84, 0.76, 0.53, 0.92, 0.83]
training_times = [1, 1, 1, 1, 1]
prediction_times = [0.92, 0.87, 1, 0.81, 0.93]

# Create subplots
fig, ax1 = plt.subplots(figsize=(12, 8))

# Plot accuracy
color = 'tab:blue'
ax1.set_xlabel('Experiment')
ax1.set_ylabel('Accuracy', color=color)
ax1.plot(experiments, accuracies, marker='o', color=color, label='Accuracy')
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_ylim(0, 1)

# Create a second y-axis for training and prediction times
ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Time (seconds)', color=color)
ax2.plot(experiments, training_times, marker='x', linestyle='--', color='tab:orange', label='Training Time')
ax2.plot(experiments, prediction_times, marker='^', linestyle='--', color='tab:red', label='Prediction Time')
ax2.tick_params(axis='y', labelcolor=color)
ax2.set_ylim(0, max(training_times + prediction_times) * 1.1)

# Add title and legend
plt.title('QSVM Experiment Results')
fig.tight_layout()
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for QSVm
qsvm_accuracies = [0.84, 0.76, 0.53, 0.92, 0.83]
qsvm_training_times = [1, 1, 1, 1, 1]
qsvm_prediction_times = [0.92, 0.87, 1, 0.81, 0.93]

# Data for SVM
svm_accuracies = np.array([0.80, 0.73, 0.52, 0.44, 0.46])
svm_training_times = np.array([0.0053, 0.0036, 0.0071, 0.0142, 0.0255])
svm_prediction_times = np.array([0.0007, 0.0008, 0.0021, 0.0028, 0.0063])

# Create subplots
fig, axs = plt.subplots(3, 1, figsize=(10, 15))

# Plot accuracies
axs[0].hist([qsvm_accuracies, svm_accuracies], bins=np.arange(0.4, 1.0, 0.1), label=['QSVm', 'SVM'], alpha=0.7)
axs[0].set_title('Comparison of Accuracies')
axs[0].set_xlabel('Accuracy')
axs[0].set_ylabel('Frequency')
axs[0].legend()

# Plot training times
axs[1].hist([qsvm_training_times, svm_training_times], bins=10, label=['QSVm', 'SVM'], alpha=0.7)
axs[1].set_title('Comparison of Training Times')
axs[1].set_xlabel('Training Time (seconds)')
axs[1].set_ylabel('Frequency')
axs[1].legend()

# Plot prediction times
axs[2].hist([qsvm_prediction_times, svm_prediction_times], bins=10, label=['QSVm', 'SVM'], alpha=0.7)
axs[2].set_title('Comparison of Prediction Times')
axs[2].set_xlabel('Prediction Time (seconds)')
axs[2].set_ylabel('Frequency')
axs[2].legend()

# Adjust layout
plt.tight_layout()
plt.show()
```
